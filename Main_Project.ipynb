# **Attributes and Description**


| S.No | Attribute                                             | Description                                                                            |
| ---- | ----------------------------------------------------- | -------------------------------------------------------------------------------------- |
| 1    | **VIN (Vehicle Identification Number)(1-10)**         | First 10 characters of the vehicle's VIN (unique vehicle identifier).                  |
| 2    | **County**                                            | County where the vehicle is registered.                                                |
| 3    | **City**                                              | City where the vehicle is registered.                                                  |
| 4    | **State**                                             | State of vehicle registration.                                                         |
| 5    | **Postal Code**                                       | ZIP/Postal code of the registered location.                                            |
| 6    | **Model Year**                                        | Manufacturing year of the electric vehicle.                                            |
| 7    | **Manufacturer**                                      | Company that manufactured the vehicle (e.g., Tesla, Nissan).                           |
| 8    | **Model**                                             | Model name of the electric vehicle.                                                    |
| 9    | **Electric Vehicle Type**                             | Type of EV (Battery Electric Vehicle (BEV) or Plug-in Hybrid Electric Vehicle (PHEV)). |
| 10   | **Clean Alternative Fuel Vehicle (CAFV) Eligibility** | Indicates whether the vehicle qualifies for CAFV incentives.                           |
| 11   | **Electric Range**                                    | Maximum distance the vehicle can travel on electric power (in miles).                  |
| 12   | **Legislative District**                              | Legislative district where the vehicle is registered.                                  |
| 13   | **DOL Vehicle ID**                                    | Unique vehicle identification number assigned by the Department of Licensing.          |
| 14   | **Vehicle Location**                                  | Geographic coordinates or registered location of the vehicle.                          |
| 15   | **Electric Utility**                                  | Electric utility provider serving the registered area.                                 |
| 16   | **2020 Census Tract**                                 | Census tract identifier based on the 2020 U.S. Census.                                 |
