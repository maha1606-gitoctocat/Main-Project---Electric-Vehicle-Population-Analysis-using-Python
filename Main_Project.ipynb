# **Electric Vehicle Population Analysis Using Python**

# **Business Problem**

As the number of electric vehicles continues to grow, government agencies, utility companies, and automobile manufacturers need to understand EV adoption trends across different regions. However, they often lack clear insights into:

Which counties and cities have the highest EV adoption.

Which manufacturers dominate the EV market.

How electric vehicle range has improved over different model years.

The percentage of vehicles eligible for Clean Alternative Fuel Vehicle (CAFV) incentives.

The distribution of Battery Electric Vehicles (BEVs) versus Plug-in Hybrid Electric Vehicles (PHEVs).

Which electric utility providers serve the largest number of EV owners.

1. To analyze the distribution of electric vehicles across different geographical regions.

# **Data Cleaning,Data Transformation and Feature Engineering**

```python
# Check duplicate rows
print("Duplicate Rows :", df.duplicated().sum())
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Fill Categorical Columns

cat_cols = df. select_dtypes(include='object').columns
for col in cat_cols:
   df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Fill Numerical Columns

num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
# Drop Unwanted Columns
df.drop(columns=['Census_Tract'], inplace=True, errors='ignore')

df.head()
```

```python
df.drop(columns=['Vehicle_Location'], inplace=True, errors='ignore')

df.head()
```

```python
# Rename Column Names
df.rename(columns={
    "VIN (Vehicle Identification Number)(1-10)":"VIN",
    "Clean Alternative Fuel Vehicle (CAFV) Eligibility":"CAFV_Eligibility",
    "Electric Vehicle Type":"EV_Type",
    "Electric Range":"Electric_Range",
    "Postal Code":"Postal_Code",
    "Model Year":"Model_Year",
    "Legislative District":"Legislative_District",
    "DOL Vehicle ID":"Vehicle_ID",
    "Electric Utility":"Electric_Utility",
   }, inplace=True)

df.columns
```

```python
# Replace Column Names
df["CAFV_Eligibility"] = df["CAFV_Eligibility"].replace({
    "Clean Alternative Fuel Vehicle Eligible": "Eligible",
    "Not eligible due to low battery range": "Not Eligible",
    "Eligibility unknown as battery range has not been researched": "Unknown"
})
```

```python
print(df["CAFV_Eligibility"].value_counts())
```

```python
# Convert Data Types
df['Postal_Code'] = df['Postal_Code'].astype(int)
df['Legislative_District'] = df['Legislative_District'].astype(int)

df.info()
```

# **Feature Engineering**

```python
# Vehicle Age
current_year = 2026

df['Vehicle_Age'] = current_year - df['Model_Year']
```

```python
# Range Category
df['Range_Category'] = pd.cut(
    df['Electric_Range'],
    bins=[0,100,250,1000],
    labels=['Low','Medium','High']
)
```

```python
# EV Category
df['EV_Category'] = np.where(
    df['Electric_Range']>=250,
    'Long Range',
    'Standard Range'
)
```

```python
df.isnull().sum()
```

```python
df[df["Range_Category"].isnull()]
```

```python
# impute null values
def range_category(x):
    if x == 0:
        return "Unknown"
    elif x <= 100:
        return "Short Range"
    elif x <= 250:
        return "Medium Range"
    else:
        return "Long Range"

df["Range_Category"] = df["Electric_Range"].apply(range_category)
```

# **Statistical analysis & Visualizations**

```python
# Measure of Central Tendency
selected_cols = ['Electric_Range', 'Model_Year', 'Vehicle_Age']

central_tendency = pd.DataFrame({
    'Mean': df[selected_cols].mean(),
    'Median': df[selected_cols].median(),
    'Mode': df[selected_cols].mode().iloc[0]
})

print("Measure of Central Tendency")
central_tendency
```

```python
# Visualization
plt.figure(figsize=(6,4))

ax = sns.histplot(
    data=df,
    x='Electric_Range',
    bins=20,
    kde=True,
    color='darkblue'
)

# Add labels to histogram bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=8)

plt.title("Distribution of Electric Range", fontsize=13, fontweight='bold')
plt.xlabel("Electric Range (Miles)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

**Intrepretation**

The analysis reveals that the EV market in this dataset is dominated by vehicles with lower electric ranges, while a smaller share consists of **modern long-range electric vehicles**. This pattern suggests a transition in the market from earlier, shorter-range EVs to newer models offering improved battery performance and extended driving range.

```python
# Variance and Standard Deviation
selected_cols = ['Electric_Range', 'Model_Year', 'Vehicle_Age']

variation = pd.DataFrame({
    'Variance': df[selected_cols].var(),
    'Standard Deviation': df[selected_cols].std()
})

print("Variance and Standard Deviation")
variation
```

```python
# visualization
plt.figure(figsize=(12,4))

for i, col in enumerate(selected_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f'{col}')
    plt.ylabel(col)

plt.suptitle('Box Plots of EV Numerical Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

**2. Variance and Standard Deviation**

**Interpretation**

**Electric_Range** typically shows the highest variation because different EV models offer different battery capacities and driving ranges.

**Model_Year** generally has lower variation since most vehicles belong to recent manufacturing years, reflecting the rapid growth of EV adoption.

**Vehicle_Age** usually has lower variation, indicating that most registered EVs are relatively new.

```python
# Skewness, kurtosis
selected_cols = ['Electric_Range', 'Model_Year', 'Vehicle_Age']

shape = pd.DataFrame({
    'Skewness': df[selected_cols].skew(),
    'Kurtosis': df[selected_cols].kurt()
})

print("Skewness and Kurtosis")
shape
```

```python
# visualization
plt.figure(figsize=(4,3))

shape.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Skewness and Kurtosis of EV Features")
plt.xlabel("Features")
plt.ylabel("Value")
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

**3. Interpretation of the Skewness and Kurtosis Chart**

**Eletrcic_Range** is often **positively skewed**, meaning most vehicles have moderate driving ranges while a few premium EVs offer exceptionally high ranges.

**Model_Year** generally has **negative skewness** because the dataset is concentrated on recent vehicle models, reflecting increasing EV adoption in recent years.

**Vehicle_Age** is usually **positively skewed**, showing that most registered EVs are relatively new, with fewer older vehicles.

**Higher kurtosis** in **Electric_Range** suggests the presence of some high-range vehicles that differ noticeably from the majority.

# **3. Multivariate Analysis**

**1. Manufacturer vs EV Type vs Average Electric Range**

```python
pivot = df.pivot_table(
    values='Electric_Range',
    index='Manufacturer',
    columns='EV_Type',
    aggfunc='mean'
)

# Get top manufacturers based on registration counts, similar to previous analysis
top_manufacturers = df['Manufacturer'].value_counts().head(10).index

plt.figure(figsize=(7,6))

sns.heatmap(
    pivot.loc[top_manufacturers], # Use .loc with the defined top_manufacturers
    annot=True,
    cmap='YlGnBu',
    fmt='.0f'
)

plt.title("Average Electric Range by Manufacturer and EV Type")

plt.show()
```

**3. Model Year × EV Type × Registration Count**

```python
# Visualization
plt.figure(figsize=(6,3))

trend = df.groupby(['Model_Year', 'EV_Type']).size().reset_index(name='Registrations')

sns.lineplot(
    data=trend,
    x='Model_Year',
    y='Registrations',
    hue='EV_Type',
    marker='o',
    linewidth=2.5
)

plt.title("EV Adoption Trend by Model Year and Vehicle Type", fontsize=14, fontweight='bold')
plt.xlabel("Model Year")
plt.ylabel("Number of Registrations")
plt.xticks(rotation=45)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Interpretation**

**Insight**

Recent model years (**especially 2023**) have the highest registration counts, indicating rapid EV adoption.

BEVs consistently outnumber PHEVs across model years, reinforcing the shift toward fully electric mobility.
