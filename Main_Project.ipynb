# **Electric Vehicle Population Analysis Using Python**

# **Business Problem**

As the number of electric vehicles continues to grow, government agencies, utility companies, and automobile manufacturers need to understand EV adoption trends across different regions. However, they often lack clear insights into:

Which counties and cities have the highest EV adoption.

Which manufacturers dominate the EV market.

How electric vehicle range has improved over different model years.

The percentage of vehicles eligible for Clean Alternative Fuel Vehicle (CAFV) incentives.

The distribution of Battery Electric Vehicles (BEVs) versus Plug-in Hybrid Electric Vehicles (PHEVs).

Which electric utility providers serve the largest number of EV owners.

# **Data Cleaning,Data Transformation and Feature Engineering**

```python
# Check duplicate rows
print("Duplicate Rows :", df.duplicated().sum())
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Fill Categorical Columns

cat_cols = df. select_dtypes(include='object').columns
for col in cat_cols:
   df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Fill Numerical Columns

num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
# Rename Column Names
df.rename(columns={
    "VIN (Vehicle Identification Number)(1-10)":"VIN",
    "Clean Alternative Fuel Vehicle (CAFV) Eligibility":"CAFV_Eligibility",
    "Electric Vehicle Type":"EV_Type",
    "Electric Range":"Electric_Range",
    "Postal Code":"Postal_Code",
    "Model Year":"Model_Year",
    "Legislative District":"Legislative_District",
    "DOL Vehicle ID":"Vehicle_ID",
    "Vehicle Location":"Vehicle_Location",
    "Electric Utility":"Electric_Utility",
    "2020 Census Tract":"Census_Tract"
}, inplace=True)

df.columns
```

```python
df.head(5)
```

```python
# Drop Unwanted Columns
df.drop(columns=['Census_Tract'], inplace=True, errors='ignore')

df.head()
```

```python
df.drop(columns=['Vehicle_Location'], inplace=True, errors='ignore')

df.head()
```

```python
df
```

```python
# Replace Column Names
df["CAFV_Eligibility"] = df["CAFV_Eligibility"].replace({
    "Clean Alternative Fuel Vehicle Eligible": "Eligible",
    "Not eligible due to low battery range": "Not Eligible",
    "Eligibility unknown as battery range has not been researched": "Unknown"
})
```

```python
print(df["CAFV_Eligibility"].value_counts())
```

```python
df.head(5)
```

```python
# Check Outliers
plt.figure(figsize=(10,5))

sns.boxplot(data=df[['Electric_Range']])

plt.show()
```

```python
# Convert Data Types
df['Postal_Code'] = df['Postal_Code'].astype(int)
df['Legislative_District'] = df['Legislative_District'].astype(int)

df.info()
```

# **Feature Engineering**

```python
# Vehicle Age
current_year = 2026

df['Vehicle_Age'] = current_year - df['Model_Year']
```

```python
df.head(2)
```

```python
# Range Category
df['Range_Category'] = pd.cut(
    df['Electric_Range'],
    bins=[0,100,250,1000],
    labels=['Low','Medium','High']
)
```

```python
df.head(2)
```

```python
# EV Category
df['EV_Category'] = np.where(
    df['Electric_Range']>=250,
    'Long Range',
    'Standard Range'
)
```

```python
df.head(2)
```

```python
df.head()

df.info()

df.shape
```

```python
df.isnull().sum()
```

```python
df[df["Range_Category"].isnull()]
```

```python
# impute null values
def range_category(x):
    if x == 0:
        return "Unknown"
    elif x <= 100:
        return "Short Range"
    elif x <= 250:
        return "Medium Range"
    else:
        return "Long Range"

df["Range_Category"] = df["Electric_Range"].apply(range_category)
```

## **Statistical analysis & Visualizations**

```python
# MEASURE OF CENTRAL TENDENCY + Chart
num_cols = df.select_dtypes(include=['int64','float64']).columns

central_tendency = pd.DataFrame({
    'Mean': df[num_cols].mean(),
    'Median': df[num_cols].median(),
    'Mode': df[num_cols].mode().iloc[0]
})

print(central_tendency)
```

```python
for col in num_cols:

    plt.figure(figsize=(4,5))

    sns.histplot(df[col], bins=30, kde=True)

    plt.axvline(df[col].mean(), color='red', linestyle='--',
                label=f"Mean = {df[col].mean():.2f}")

    plt.axvline(df[col].median(), color='green', linestyle='-',
                label=f"Median = {df[col].median():.2f}")

    plt.title(f"Distribution of {col}")

    plt.legend()

    plt.show()
```

**1. Histogram (Mean & Median)**

**Interpretation (Symmetric Distribution)**

The mean and median are very close, indicating the data is approximately normally distributed.
Most observations are concentrated around the center.
There is little evidence of extreme values affecting the distribution.

**Interpretation (Right-Skewed Distribution)**

The mean is greater than the median.
A few high values pull the mean toward the right.
Most observations are concentrated in the lower range.

**Interpretation (Left-Skewed Distribution)**

The mean is less than the median.
Some unusually low values pull the mean toward the left.
Most observations are concentrated in the higher range.

```python
# Variance and Standard Deviation
variation = pd.DataFrame({
    'Variance': df[num_cols].var(),
    'Standard Deviation': df[num_cols].std()
})

print(variation)
```
